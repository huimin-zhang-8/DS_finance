### 1.1 Création du dataset pour la regression 

```python
#1. Chargement des données
def chargement_des_donnees(filepaths):
    df = pd.read_csv(filepaths, usecols=['Close'])
    return df
```

```python
#2. Standardisation des données et séparation en train et test
def standardisation_des_donnees(df1,df2):
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(df1)
    data1_scaled = scaler.transform(df2)
    return data_scaled,data1_scaled, scaler

def separation_train_test(data_scaled, train_size=0.8):
    train_size = int(len(data_scaled) * train_size)
    train_data = data_scaled[:train_size]
    test_data = data_scaled[train_size:]
    return train_data, test_data
```

```python
#3 Création des datasets de features(X) et de labels (Y)

def create_target_features(df,n=30):
    x=[]
    y=[]
    for i in range(n,df.shape[0]):
        x.append(df[i-n:i,0])
        y.append(df[i,0])
    x = np.array(x)
    y = np.array(y)
    return x,y
```

```python
#4. datasets obtenus :

def datasets_regression(data):
    df = chargement_des_donnees(data)
    train, test = separation_train_test(df, train_size=0.8)
    train_scaled,test_scaled, scaler = standardisation_des_donnees(train, test)
    X_train,y_train = create_target_features(train_scaled, n=30)
    X_test,y_test = create_target_features(test_scaled, n=30)
    return X_train,y_train,X_test,y_test, scaler

```

```python
filepaths=glob.glob("/Users/chloe/Documents/M2/S4/Data Science/Cours 1/Companies_historical_data/*.csv")
```

```python
resultats_entreprises = {}

for filepath in filepaths:
    nom = os.path.basename(filepath).replace(".csv", "")
    
    X_train, X_test, y_train, y_test, scaler = datases_regression(filepath)
    
    resultats_entreprises[nom] = {
        "X_train": X_train,
        "X_test": X_test,
        "y_train": y_train,
        "y_test": y_test,
        "scaler": scaler
    }
```
